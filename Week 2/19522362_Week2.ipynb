```python
import pandas as pd
```

```python
df = pd.read_csv('sales.csv')
```

```python
df.head(5)
```

```python
# ✓ Fix column datatypes.

# Convert the "name"
df['name'] = df['name'].str.replace('"', '')

# Convert the "ordered_at" column to datetime datatype
df['ordered_at'] = pd.to_datetime(df['ordered_at'])

# Convert the "price" and "line_total" columns to float datatype
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['line_total'] = df['line_total'].replace('[\$,]', '', regex=True).astype(float)

# Print the datatypes of all columns to check if the column datatypes have been fixed
print(df.dtypes)
```

```python
# ✓ Drop if duplicated or null.
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df.head(5)
```

```python
# ✓ Sanity check for value ranges and to check assumptions.
# Check value ranges for the "quantity" column
print(df['quantity'].describe())
```

```python
# Check value ranges for the "price" column
print(df['price'].describe())
```

```python
# Check value ranges for the "line_total" column
print(df['line_total'].describe())
```

```python
# Check assumptions about the "name" column
print(df['name'].value_counts())
```

```python
# Check assumptions about the "ordered_at" column
print(df['ordered_at'].min())
print(df['ordered_at'].max())
```

```python
# ✓ Use regular expression and lambda function to parse data.
import re

# Define regular expressions for category, flavor, and topping
category_regex = r'^"(.*?)"'
flavor_regex = r'(?<=")(?:[^"]|(?<=\\)")+(?=")'
topping_regex = r'^"CONE".*?"(.*?)"$'

# Apply lambda functions to extract category, flavor, and topping information
df['category'] = df['name'].apply(lambda x: re.findall(category_regex, x)[0] if re.findall(category_regex, x) else None)
df['flavor'] = df['name'].apply(lambda x: re.findall(flavor_regex, x)[0] if re.findall(flavor_regex, x) else None)
df['topping'] = df['name'].apply(lambda x: re.findall(topping_regex, x)[0] if re.findall(topping_regex, x) and 'CONE' in x else None)

# Print the first 5 rows of the DataFrame to check if the data has been parsed correctly
df.head(5)
```

```python
import pandas as pd
```

```python
# Load the data using Pandas.
data = pd.read_csv('job-market.csv')
```

```python
# Visualize top 10 first rows
data.head(10)
```

```python
data.info()
```

```python
# Clean the data.
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
# Fix column datatypes.
import re

# Define a function to clean a string
def clean_string(text):
    if text:
        # Remove HTML tags
        clean = re.compile('<.*?>')
        text = re.sub(clean, '', text)
        
        # Replace '&nbsp;' and '\n *&nbsp;' with a space
        text = re.sub(r'(&nbsp;|\n *&nbsp;)', ' ', text)
        
        # Remove unwanted characters
        text = re.sub(r'\n *|\*', '', text)
        
        return text if text else None

data['Id'] = data['Id'].astype(int)
data['LowestSalary'] = data['LowestSalary'].astype(int)
data['HighestSalary'] = data['HighestSalary'].astype(int)
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%dT%H:%M:%S.%fZ')
data['LowestSalary'] = pd.to_numeric(data['LowestSalary'], errors='coerce')
data['HighestSalary'] = pd.to_numeric(data['HighestSalary'], errors='coerce')
data['FullDescription'] = data['FullDescription'].apply(clean_string)
```

```python
# Check and clean the data.
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```
